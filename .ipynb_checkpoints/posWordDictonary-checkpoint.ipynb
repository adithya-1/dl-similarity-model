```python
import pandas as pd
```

```python
df=pd.read_csv('dataset.csv')
```

```python
a=df['Mixed'].to_list()
```

```python
idi=df['ID'].to_list()
```

```python
b=df['POS'].to_list()
```

```python

for i in range(len(a)):
    print(idi[i],a[i])
    suba=a[i].split()
    subb=b[i].split()
    if(len(suba)!=len(subb)):
        print(idi[i],a[i],b[i])
```

```python
d=dict()
for i in range(len(a)):
    suba=a[i].split()
    subb=b[i].split()
    for j in range(len(suba)):
        if suba[j] not in d.keys():
            d[suba[j]]=[subb[j]]
        else:
            if subb[j] not in d[suba[j]]:
                d[suba[j]].append(subb[j])
                
```

```python
dSub={}
for k in sorted(d.keys()):
    if(len(d[k])>1):
        dSub[k]=d[k]
```

```python
e=dict()
```

```python
for i in range(len(a)):
    suba=a[i].split()
    subb=b[i].split()
    for j in range(len(suba)):
        if suba[j] in dSub.keys():
            if suba[j] not in e.keys():
                e[suba[j]]=[subb[j],idi[i]]
            else:
                e[suba[j]].append([subb[j],idi[i]])
                
```
