```python
import pandas as pd
```

```python
df=pd.read_csv('DL_sample_sentences.csv')
```

```python
a=df['Mixed'].to_list()
```

```python
b=df['POS'].to_list()
```

```python
a=[i.split() for i in a]
```

```python
b=[i.split() for i in b]
```

```python
b
```

```python
dic=dict()
for i in range(len(b)):
    for j in range(len(b[i])):
        if b[i][j] not in dic:
            dic[b[i][j]]=[]
        if a[i][j] not in dic[b[i][j]]:
            dic[b[i][j]].append(a[i][j])
```

```python
import pickle
 

 
# Step 2
with open('posToWord.dictionary', 'wb') as config_dictionary_file:
 
  # Step 3
  pickle.dump(dic, config_dictionary_file)
```
